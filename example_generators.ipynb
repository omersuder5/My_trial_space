```python
import numpy as np
import torch 
import matplotlib.pyplot as plt

import utils.data
from generators.generators import *

%load_ext autoreload
%autoreload 2

device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
dtype=torch.float64
```

# AR

```python
T = 200 # length of time series
N = 50  # number of samples
d = 1   # dimension of time series
p = 2   # AR order
q = 0   # MA order
phi = [0.7, -0.2] # AR coefficients
theta = None # MA coefficients

x_ar = ARMA(T=T, p=p, q=q, phi=phi, theta=theta)
```

### normal noise

```python
x_ar_samples = x_ar.generate(N=N, noise=Noise("normal")).to(device=device, dtype=dtype)

plt.figure(figsize=(10, 3))
for idx in range(5):
    plt.plot(x_ar_samples[idx].squeeze().numpy(), label=f'Sample {idx+1}')
plt.title('First 5 AR Samples with Normal Noise')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

### t noise

```python
x_ar_samples = x_ar.generate(N=N, noise=Noise("t")).to(device=device, dtype=dtype)

plt.figure(figsize=(10, 3))
for idx in range(5):
    plt.plot(x_ar_samples[idx].squeeze().numpy(), label=f'Sample {idx+1}')
plt.title('First 5 AR Samples with t-Distributed Noise')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

### gamma noise

```python
x_ar_samples = x_ar.generate(N=N, noise=Noise("gamma")).to(device=device, dtype=dtype)

plt.figure(figsize=(10, 3))
for idx in range(5):
    plt.plot(x_ar_samples[idx].squeeze().numpy(), label=f'Sample {idx+1}')
plt.title('First 5 AR Samples with Gamma-Distributed Noise')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```
